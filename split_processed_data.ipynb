```
import os
import time
import json
import copy 

import numpy as np
# from sklearn.model_selection import train_test_split
```

```
list_classes = ["Containment and Closure Policies", "Economic Policies", "Health System Policies", "Miscellaneous Policies"]
dataset_path = os.path.join(os.getcwd(), "dataset")
```

```
data_path = os.path.join(dataset_path, "processed.json")
with open(data_path) as f:
    data_list = json.load(f)
```

```
data_list[0]
```

```
data_seperated_by_classes = {}
for class_name in list_classes:
    data_seperated_by_classes[class_name] = []
```

```
for idx, data in enumerate(data_list):
    if 'choices' not in data.keys(): continue
    choices = data['choices']

    for choice in choices:
        data_only_one_choice = copy.deepcopy(data)
        del data_only_one_choice['choices']
        data_only_one_choice['choice'] = choice
        data_seperated_by_classes[choice].append(data_only_one_choice) 
```

```
for class_name in list_classes:
    print(f"# of data points in {class_name}: {len(data_seperated_by_classes[class_name])}")
```

```
NUM_DATA_POINTS = 30
```

```
train_data = []
test_data = []

for class_name in list_classes:
    data_of_a_class = data_seperated_by_classes[class_name]
    randIdx = np.random.choice(len(data_of_a_class), NUM_DATA_POINTS, replace=False).tolist()

    # Train data
    selected_data = [data_of_a_class[idx] for idx in randIdx]
    train_data += selected_data

    # Test data
    selected_data = [data_of_a_class[idx] for idx in range(len(data_of_a_class)) if idx not in randIdx]
    test_data += selected_data
```

```
print(f"# of data points in training data: {len(train_data)}")
print(f"# of data points in testing data: {len(test_data)}")
```
